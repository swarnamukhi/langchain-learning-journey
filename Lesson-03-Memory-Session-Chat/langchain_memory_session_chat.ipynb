```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    google_api_key="GOOGLE-API-KEY")
```

```python
store = {}
```

```python
def get_session_history(session_id):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
```

```python
chain = RunnableWithMessageHistory(model, get_session_history)
```

```python
session_id = "1209"
```

```python
from IPython.display import Markdown


# First message
response1=chain.invoke("Hi", config={"configurable": {"session_id": session_id}})
Markdown(response1.content)

```

```python
response2=chain.invoke("Suggest places for travelling in summer", config={"configurable": {"session_id": session_id}})
Markdown(response2.content)
```

```python
response3=chain.invoke("Which country is cheapest to travel from India? Also give its currency value in INR.", config={"configurable": {"session_id": session_id}})
Markdown(response3.content)
```
