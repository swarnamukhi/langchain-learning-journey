```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    google_api_key="YOUR GOOGLE API"
)
```

```python
messages=[SystemMessage(content="You are a helpful assistant"),
          HumanMessage(content="Explain importance of Pyhsical exercise in simple terms")]
```

```python
helpfull_model = llm.bind(temperature=0, max_output_tokens=1000)
```

```python
response = helpfull_model.invoke(messages)
```

```python
from IPython.display import Markdown
Markdown(response.text)
```
